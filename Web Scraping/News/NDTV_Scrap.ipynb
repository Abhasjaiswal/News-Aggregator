```python
#test
r = requests.get("https://www.ndtv.com/tamil-nadu-news/after-supreme-court-relief-disqualified-tamil-nadu-minister-and-dmk-leader-k-ponmudy-to-get-post-back-5231790")
soup = BeautifulSoup(r.text, 'lxml')

for p_tag in soup.find_all('p'):
    print(p_tag.text.strip())

import spacy
nlp = spacy.load("en_core_web_sm")
paragraphs = [p.text.strip() for p in soup.find_all('p')]
text = "\n".join(paragraphs)
doc = nlp(text)
summary = "\n".join([sent.text for sent in doc.sents])
print(summary)
```
