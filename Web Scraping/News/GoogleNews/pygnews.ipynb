```python
gn=GoogleNews()
```

```python
top = gn.top_news()
top
```

```python
for entry in top['entries']:
    title = entry['title']
    link = entry['links'][0]['href']
    published = entry['published']
    summary = entry['summary']
    print("Title:", title)
    print("Link:", link)
    print("Published:", published)
    print("Summary:", summary)
    print()
```
